```python
#@title Importing the csv using pandas

# Importing pandas
import pandas as pd
import numpy

# Reading the contents of the file
filepath = r"/content/raw_titanic_data.csv";
titanic_data = pd.read_csv(filepath)

# Getting the row count and the coloumn count to doublecheck if all the data has been processed
row_count = len(list(titanic_data.axes[0]))
column_count = len(list(titanic_data.axes[1]))
# print("Row count: " + str(row_count))
# print("Column count: " + str(column_count))

# Printing the data
titanic_data.head()
```

```python
#@title [1] Retrieve the first 10 rows of the dataset.

# Getting the first 10 lines of titanic_data
titanic_data.head(10)
```

```python
#@title [2] Display the columns Name, Age, and Survived for passengers with a Pclass value of 1.

# Filtering data for passengers with Pclass = 1
pclass_1_passengers = titanic_data[titanic_data['Pclass'] == 1]

# Selecting the desired columns
selected_columns = pclass_1_passengers[['Name', 'Age', 'Survived']]

# Showing the filtered data
display(selected_columns)
```

```python
#@title [3] Find and display all rows where passengers are aged above 60.

# Filtering the Age to find people older than 60
passengers_over_60 = titanic_data[titanic_data['Age'] > 60]

# Showing the filtered data
display(passengers_over_60)
```

```python
#@title [4] Locate the row(s) for the passenger(s) with the maximum fare (Fare)

# Getting the maximum fare
max_fare = titanic_data['Fare'].max()

# Filtering the data to find rows with the maximum fare
passengers_with_max_fare = titanic_data[titanic_data['Fare'] == max_fare]

# Showing the filtered data
display(passengers_with_max_fare)
```

```python
#@title [5] Retrieve the Name and Ticket of passengers who embarked from port C (Cherbourg).

# Getting the passengers who embarked at Cherbourg
cherbourg_passengers = titanic_data[titanic_data['Embarked'] == 'C']

# Getting their Name and Ticket
name_ticket = cherbourg_passengers[['Name', 'Ticket']]

# Showing the filtered data
display(name_ticket)
```

```python
#@title [6] Display the Name, Age, and Cabin for passengers who do not have a cabin assigned (where Cabin is NaN).

# Getting the passengers who have not been assigned a cabin
nocabin_passengers = titanic_data[titanic_data['Cabin'].isnull()]

# Getting their Name, Age, and Cabin
name_age_cabin = nocabin_passengers[['Name', 'Age', 'Cabin']]

# Showing the filtered data
display(name_age_cabin)
```

```python
#@title[7] Retrieve the last 5 rows of the dataset.

# Getting the last 5 rows
titanic_data.tail(5)
```

```python
#@title [8] Locate the rows where the Survived value is 0 and display the PassengerId, Name, and Sex.

# Getting the passengers who did not survive
not_survived_passengers = titanic_data[titanic_data['Survived'] == 0]

# Getting their Passenger Id, Name, and Gender
id_name_sex = not_survived_passengers[['Id', 'Name', 'Sex']]

# Showing the filtered data
display(id_name_sex)
```

```python
#@title [9] Extract all unique values of the Embarked column.

# Getting all the unique values of the Embarked coloumn
titanic_data['Embarked'].unique()
```

```python
#@title [10] Select the rows of passengers with a Fare greater than 50 and sort them in descending order by Fare.

# Filtering the Fare to find values above 50
fare_over_50 = titanic_data[titanic_data['Fare'] > 50]

# Arranging the dataframe in descending order
fare_over_50 = fare_over_50.sort_values('Fare', ascending=False)

# Showing the filtered data
display(fare_over_50)
```
