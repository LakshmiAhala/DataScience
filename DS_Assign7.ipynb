```python
#@title Importing the .csv using pandas

# Importing pandas
import pandas as pd
import numpy

# Reading the contents of the file
filepath = r"/content/raw_titanic_data.csv";
titanic_data = pd.read_csv(filepath)

# Printing the data
titanic_data.head()
```

```python
#@title Task 2: Perform GroupBy analysis in Pandas

from typing import final

# Group data by 'Pclass'
grouped_data = titanic_data.groupby('Pclass')

# Survivors count per passenger class
surv_count_by_class = grouped_data['Survived'].count()
# print('Count of Survivors per Passenger Class: ')
# print(surv_count_by_class)

# Average fare per passenger class
avg_fare_by_class = grouped_data['Fare'].mean()
avg_fare_by_class = round(avg_fare_by_class, 2)
# print('\nAverage Fare per Passenger Class: ')
# print(avg_fare_by_class)

# Average age per passenger class
avg_age_by_class = grouped_data['Age'].mean()
avg_age_by_class = round(avg_age_by_class, 2)
# print('\nAverage Age per Passenger Class:')
# print(avg_age_by_class)

# Creating a dictionary with the data to form a dataframe
final_data = {
    'Passenger Class': surv_count_by_class.index,
    'Survivors Count': surv_count_by_class.values,
    'Average Fare': avg_fare_by_class.values,
    'Average Age': avg_age_by_class.values
}

# Create the DataFrame
final_df = pd.DataFrame(final_data)

# Display the DataFrame
final_df.head()
```

```python
#@title Task 3: Perform Pivot Table Analysis Using Pandas

import pandas as pd

pivot_table = pd.pivot_table(
                       titanic_data,
                       values=['Survived', 'Fare', 'Age'],
                       index=['Pclass'],
                       aggfunc={'Survived': 'count', 'Fare': 'mean', 'Age': 'mean'}
)

pivot_table = pivot_table.rename(
    columns={
        'Survived': 'Survivors Count',
        'Fare': 'Average Fare',
        'Age': 'Average Age'
    }
)
print(pivot_table)

```
